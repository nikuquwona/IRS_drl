```python
import matplotlib.pyplot as plt
import numpy as np

# 生成示例数据
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.exp(-x)

# 创建图形对象和主轴
fig, ax1 = plt.subplots()

# 绘制第一条折线
ax1.plot(x, y1, color='blue')

# 设置主轴标题和纵轴标签
ax1.set_title('Example with Two Y-Axes')
ax1.set_ylabel('Sine Wave', color='blue')

# 创建第二个纵轴并绑定到主轴上
ax2 = ax1.twinx()

# 绘制第二条折线
ax2.plot(x, y2, color='red')

# 设置第二个纵轴标签
ax2.set_ylabel('Exponential Decay', color='red')

# 自动调整 x 轴范围
ax1.autoscale()

# 显示图形
plt.legend()
plt.show()
```
